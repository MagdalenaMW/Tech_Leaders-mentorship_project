```python
df5_1990 = df5[df5["Year"] == 1990].copy()

def categorize_hdi(hdi):
    if hdi <= 0.3499:
        return "Very Low"
    elif hdi <= 0.5499:
        return "Low"
    elif hdi <= 0.6999:
        return "Medium"
    elif hdi <= 0.7999:
        return "High"
    else:
        return "Very High"

df5_1990["HDI Category"] = df5_1990["Human Development Index (UNDP)"].apply(categorize_hdi)


HDI_group_counts = df5_1990.groupby("HDI Category")["Entity"].count()

custom_order = ["Very Low", "Low", "Medium", "High", "Very High"]
HDI_group_counts = HDI_group_counts.reindex(custom_order, fill_value=0)

colors = {
    "Very Low": "forestgreen",
    "Low": "darkorange",
    "Medium": "indianred",
    "High": "steelblue",
    "Very High": "lightseagreen"
}

plt.figure(figsize=(4, 5))
bars = plt.bar(HDI_group_counts.index,
               HDI_group_counts.values,
               color=[colors[cat] for cat in HDI_group_counts.index])
plt.title("Number of Countries by HDI Category in 1990")
plt.grid(axis = "y",  linestyle = "--", color = "lightgray")
plt.xlabel("HDI Category")
plt.ylabel("Number of Countries")

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, f'{yval:.0f}', ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
colors = {
    "Poland": "blue",
    "India": "orange",
    "Germany": "green",
    "United States": "red",
    "South Africa": "purple",
    "Iceland": "brown"
}

countries = list(colors.keys())
emission_types = ["bc", "oc", "so2", "co", "nox", "nmvoc", "nh3"]
years = sorted(df1["Year"].unique())

for typ in emission_types:
    plt.figure(figsize=(10, 6))

    for country in countries:
        emission = df1[df1["Entity"] == country].sort_values("Year")

        column = [col for col in df1.columns if col.endswith(f"_{typ}")]

        emission_sum = emission[column].sum(axis=1)

        plt.plot(emission["Year"], emission_sum, label=country, color=colors[country])

    plt.title(f"Emission type: {typ.upper()}")
    plt.xlabel("Year")
    plt.ylabel("Emission [t]")
    plt.grid(True, linestyle="--", alpha=0.5)
    plt.legend()
    plt.tight_layout()
    plt.show()
```
